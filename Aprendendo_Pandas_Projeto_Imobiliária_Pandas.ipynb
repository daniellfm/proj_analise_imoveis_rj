Qual o percentual de cada tipo de imóvel na nossa base de dados?


```python
df.Tipo.unique()
# usar o método value_counts para contar valores, lembre-se de usar 'Tipo'.
```

```python
df.Tipo.value_counts()
# diz quantas vezes o tipo de imóvel aparece na base de dados da imobiliaria
```

```python
# mas qual é o percentual
df.Tipo.value_counts(normalize=True)
# para melhorar a visualização é preciso transforma essa series em um dataframe. Para isso, usa-se o método .toframe().
```

```python
df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
# os percentuais estão ordenados do menor para o maior, o método sort_values ordena os tipo da tabela, equivalente ao order by do SQL.
```

```python
df_percentual_tipo = df['Tipo'].value_counts(normalize=True).to_frame()
df_percentual_tipo.columns = ['Percentual']  # renomear a coluna para facilitar
df_percentual_tipo = df_percentual_tipo.sort_values('Percentual')  # ordenar do menor para o maior

df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color='green', edgecolor='black',
                        xlabel='Tipos', ylabel='Percentual');
```

#Manipulando Dados
Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados.fillna(0,inplace = True)
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
# É possível criar colunas numéricas e operar matemáticamente para encontrar resultados desejados em coluna
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

Criando colunas categóricas


```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
#Salvando dados em arquivos csv
dados.to_csv('dados_completos.csv',index=False, sep =';')
```
